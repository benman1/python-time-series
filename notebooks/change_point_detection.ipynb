# Synthetic dataset

```python
import numpy as np
import matplotlib.pylab as plt
import ruptures as rpt
signal, bkps = rpt.pw_constant(
  n_samples=500, n_features=3, n_bkps=3,
  noise_std=5.0, delta=(1, 20)
)
rpt.display(signal, bkps)
```

```python
algo = rpt.Binseg(model="l1").fit(signal)
my_bkps = algo.predict(n_bkps=3)
rpt.show.display(signal, bkps, my_bkps, figsize=(10, 6))
```

# Amazon stock

```python
import yfinance as yf

data = yf.download('AMZN', start='2010-01-01', end='2021-12-31')
prices = data['Adj Close'].values
```

```python
import ruptures as rpt

model = rpt.Pelt(model="rbf").fit(prices)
result = model.predict(pen=10)
```

```python
import matplotlib.pyplot as plt

plt.plot(prices)
for i in result:
    plt.axvline(i, color='red', linestyle='--')
plt.show()
```
